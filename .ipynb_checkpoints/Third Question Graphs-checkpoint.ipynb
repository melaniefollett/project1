```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats
from pprint import pprint
```

```python
#Import csv
raw_game_data = pd.read_csv("vgsales-12-4-2019.csv", encoding="UTF-8")
raw_game_data.head()
```

```python
#Filter columns to only ones we need for analysis
base_game_df = raw_game_data.loc[:, ["Name", "Genre", "Platform", "Critic_Score", "Global_Sales", "NA_Sales",
                                     "PAL_Sales", "JP_Sales", "Other_Sales", "Year"]]

#Rename columns
base_game_df = base_game_df.rename(columns={"Critic_Score": "Critic Score", "Global_Sales": "Global Sales", "NA_Sales": "NA Sales",
                                           "PAL_Sales": "PAL Sales", "JP_Sales": "JP Sales", "Other_Sales": "Other Sales"})

base_game_df
```

```python
#Filter Year to 2009-2019
game_df = base_game_df[(base_game_df['Year'] >= 2013) & (base_game_df['Year'] <= 2017)]
#game_df = base_game_df[base_game_df["Year"] >= 2014]
game_df
```

```python
game_df2 = game_df.loc[:, ["Genre", "Global Sales", "Year"]]
game_df2.head()

#drop NaN's
game_df3 = game_df2.dropna()
game_df3
```

```python
game_df3 = game_df3.replace({"Action": "Action Adventure", "Action-Adventure": "Action Adventure", "Adventure": "Action Adventure", "Shooter": "Action Adventure",
    "Role-Playing": "MMO/Role-Playing", "MMO": "MMO/Role-Playing", "Visual Novel": "MMO/Role-Playing",
    "Sports": "Sports/Simulation", "Simulation": "Sports/Simulation", "Racing": "Sports/Simulation",
    "Puzzle": "Puzzle/Party", "Party": "Puzzle/Party", "Music": "Puzzle/Party", "Board Game": "Puzzle/Party", "Strategy": "Puzzle/Party",
    "Education": "MISC", "Platform": "MISC", "Misc": "MISC"})
game_df3
```

```python
# Group genres together
game_year = game_df3.groupby(['Genre', 'Year'])
# Get the total global sales per year
yearly_sales = game_year['Global Sales'].sum()
# Chart the Data

yearly_sales
```

```python
genre_sales = pd.DataFrame({
    "Global Sales": yearly_sales
})
genre_sales = pd.DataFrame(genre_sales).reset_index(['Genre', 'Year'])
genre_sales
```

```python
genre_sales = genre_sales.pivot(index="Year", columns="Genre", values="Global Sales")
genre_sales
```

```python
genre_sales = genre_sales.reset_index()
genre_sales
```

```python
#genre_sales.groupby("Genre").plot(y="Global Sales", figsize=(10,5), marker='.', markersize=10, title='Genre Global Sales by Year')


plt.figure(figsize=(20,20))
plt.plot(genre_sales['Year'], genre_sales['Action Adventure'])
plt.plot(genre_sales['Year'], genre_sales['Fighting'])
plt.plot(genre_sales['Year'], genre_sales['MISC'])
plt.plot(genre_sales['Year'], genre_sales['MMO/Role-Playing'])
plt.plot(genre_sales['Year'], genre_sales['Puzzle/Party'])
plt.plot(genre_sales['Year'], genre_sales['Sports/Simulation'])


plt.title('Genre Global Sales by Year', fontsize=36)
plt.xlabel('Year', fontsize=24)
plt.ylabel('Global Sales (m)', fontsize=24)
plt.legend(['Action Adventure', 'Fighting', 'MISC', 'MMO/Role-Playing', 'Puzzle/Party', 'Sports/Simulation'],loc="upper right", prop={'size': 20})
plt.show
plt.tight_layout()
plt.savefig('../Genre_Global_sa.png')
```
